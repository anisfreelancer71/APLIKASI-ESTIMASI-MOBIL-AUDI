# Visualisasi Data

```python
# Menghapus kolom yang mengandung nilai non-numerik
df_audi = audi.select_dtypes(include=[np.number])

plt.figure(figsize=(12,10))
sns.heatmap(df_audi.corr(), annot=True)
```

# Jumlah mobil berdasarkan model

```python
models = audi.groupby('model').count()[['tax']].sort_values(by='tax', ascending=True).reset_index()
models = models.rename(columns={'tax':'numberofcar'})


fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['model'], y=models['numberofcar'], color='royalblue')
plt.xticks(rotation=60)
```

# Ukuran mesin

```python
engine = audi.groupby('engineSize').count()[['tax']].sort_values(by='tax').reset_index()
engine = engine.rename(columns={'tax':'count'})

plt.figure(figsize=(15,5))
sns.barplot(x=engine['engineSize'], y=engine['count'], color='royalblue')
```

# Distribusi mileage

```python
plt.figure(figsize=(15,5))
sns.histplot(df_audi['mileage'])
```

# Distribusi harga mobil

```python
features = ['year','mileage','tax','mpg','engineSize']
x = df_audi[features]
y = df_audi['price']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x, y, random_state=70)
y_test.shape
```

# Membuat model regresi linear

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
```

# Akurasi Score

```python
score = lr.score(x_test, y_test)
print('Akurasi model Regresi', score)
```

# Membuat inputan regresi linear

```python
# input_data year=2019, mileage=5000, tax=145, mpg=30.2, engineSize=2
input_data = np.array([[2019,5000,145,30.2,2]])

prediction = lr.predict(input_data)
print('Estimasi harga mobil Audi dalam EUR :', prediction)
```
